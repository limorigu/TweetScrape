```python
import requests

url = "https://twitter135.p.rapidapi.com/v2/UserTweets/"

querystring = {
    # TODO: specify user_id of target twitter handle
    "id":"",
    "count":"10",
    "start_date": "2023-10-07T04:30:00+00:00",
    "end_date": "2023-10-07T20:00:00+00:00"}

headers = {
    # TODO: replace with your RapidAPI key
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "twitter135.p.rapidapi.com",
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
# format of response into pretty json
import json
print(json.dumps(response.json(), indent=4, sort_keys=True, ensure_ascii=False))
```

```python
response.json()['data']['user']['result']['timeline_v2']['timeline']['instructions']
```

```python
for entry in response.json()['data']['user']['result']['timeline_v2']['timeline']['instructions'][2]['entries']:
    if 'who-to-follow' in entry['entryId']:
        continue
    else:
        try:
            post = entry['content']['itemContent']['tweet_results']['result']['legacy']
            if post['is_quote_status']:
                print("retweet with comment")
                print("1")
                print(post['full_text'])
                print()
                print(json.dumps(
                    entry['content']['itemContent']['tweet_results']['result']['quoted_status_result']['result']['legacy']['full_text'], 
                         indent=4, sort_keys=True, ensure_ascii=False))
                print(json.dumps(
                            entry['content']['itemContent']['tweet_results']['result']['quoted_status_result']['result']['core']['user_results']['result']['legacy']['name'], 
                         indent=4, sort_keys=True, ensure_ascii=False))
                print("done with retweet")
                print(" ")
        except KeyError:
            try:
                print()
                print("multi-comment")
                print("2")
                for item in entry['content']['items']:
                    post = item['item']['itemContent']['tweet_results']['result']['legacy']
                    if post['is_quote_status']:
                        print("retweet with comment")
                        print(post['full_text'])
                        # ['itemContent']['tweet_results']['result']['quoted_status_result']['legacy']['full_text']
                        print(json.dumps(
                            item['item']['itemContent']['tweet_results']['result']['quoted_status_result']['result']['legacy']['full_text'], 
                                indent=4, sort_keys=True, ensure_ascii=False))
                        print(json.dumps(
                                    item['item']['itemContent']['tweet_results']['result']['quoted_status_result']['result']['core']['user_results']['result']['legacy']['name'], 
                                indent=4, sort_keys=True, ensure_ascii=False))
                        print("done with retweet")
                        print(" ")
                    else:
                        print("tweet")
                        print(json.dumps(post['full_text'], indent=4, sort_keys=True, ensure_ascii=False))
            except KeyError:
                print(json.dumps(entry, indent=4, sort_keys=True, ensure_ascii=False))
```
